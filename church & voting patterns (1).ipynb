Religion and Politics: Uganda held a presidential election in January 2021. In this notebook, I analyse relationship between politics and regions by looking at how polling stations that were situated at religious institutions voted. I found out that Uganda's pop-star turned musician Robert Kyagulanyi aka Bobi Wine performed better at polling stations that were located at Muslim related institutions. <a href="https://blanshe.medium.com/bobi-wine-performed-better-at-muslim-institutions-polling-stations-3158284ff82c">And here is a story I wrote from it</a>
